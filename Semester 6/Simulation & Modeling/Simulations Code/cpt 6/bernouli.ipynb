###  Write a program to implement Bernouli distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 0.6
n_samples = 1000

bernoulli_samples = np.random.binomial(1, p, n_samples)

print("Bernoulli First 10 Samples:", bernoulli_samples[:10])
print("Bernoulli Empirical Mean:", np.mean(bernoulli_samples))
print("Bernoulli Theoretical Mean:", p)

# Plot Bernoulli
plt.hist(bernoulli_samples, bins=2, edgecolor='black', rwidth=0.8)
plt.title("Bernoulli Distribution (n=1)")
plt.xticks([0, 1])
plt.show()


```

## Lab 04: Bernoulli Distribution — Quality-Control Test for Microchips

**Problem:** A Manufacturing and Reliability Lab tests microchips. Each chip either works (Success = 1) or fails (Failure = 0). Historical data shows **p = 0.85** (probability of passing).

| X | Meaning |
|---|---------|
| 1 | Chip passes the test (Success) |
| 0 | Chip fails the test (Defect) |

**Theoretical Results:**
- P(X = 0) — Probability of defect = 1 − p = 0.15
- E[X] — Expected value (Mean) = p = 0.85
- Var(X) — Variance = p(1 − p) = 0.85 × 0.15 = 0.1275

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 0.85          # probability of chip passing
n_samples = 1000

# Simulate Bernoulli trials (1 = pass, 0 = fail)
samples = np.random.binomial(1, p, n_samples)

# ── Analytical Results ──────────────────────────────────────────
p_defect   = 1 - p             # P(X = 0)
mean_th    = p                  # E[X]
var_th     = p * (1 - p)       # Var(X)

# ── Empirical Results ───────────────────────────────────────────
mean_emp   = np.mean(samples)
var_emp    = np.var(samples)

print("=" * 45)
print("   Microchip Quality-Control — Bernoulli")
print("=" * 45)
print(f"  p (pass probability)     : {p}")
print(f"  P(X=0) — Defect prob     : {p_defect:.4f}")
print(f"  E[X]   — Theoretical mean: {mean_th:.4f}")
print(f"  Var(X) — Theoretical var : {var_th:.4f}")
print("-" * 45)
print(f"  Empirical Mean           : {mean_emp:.4f}")
print(f"  Empirical Variance       : {var_emp:.4f}")
print("=" * 45)

# ── Plot ────────────────────────────────────────────────────────
counts = [np.sum(samples == 0), np.sum(samples == 1)]
bars = plt.bar([0, 1], counts, color=['#e74c3c', '#2ecc71'],
               edgecolor='black', width=0.4)

for bar, cnt in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 10,
             f'{cnt}', ha='center', fontsize=11, fontweight='bold')

plt.title("Bernoulli Distribution — Microchip Quality Control (p = 0.85)", fontsize=13)
plt.xlabel("Outcome  (0 = Fail / Defect,  1 = Pass / Success)")
plt.ylabel("Frequency")
plt.xticks([0, 1], ['0 — Fail (Defect)', '1 — Pass (Success)'])
plt.ylim(0, max(counts) + 80)
plt.tight_layout()
plt.show()
```
