### Lab 04: Bernoulli data distribution: A Manufacturing and Reliability Lab is testing a newly produced batch
of microchips. Each chip undergoes a strict operational test and either works perfectly (Success = 1) or fails
(Failure = 0). Historical data shows that the probability of any given chip passing this test is p = 0.85. This lab
problem models a Bernoulli distribution using a Quality-Control Test. It guides students through defining a
binary outcome (success/failure). State the values of X and what they represent. Calculating probabilities, and
finding expected metrics. Calculate the probability of a defect, P(X=0). Find the mean: Calculate the expected
value, E[X]. Find the spread: Calculate the variance, Var(X).

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.85         
n_samples = 1000

samples = np.random.binomial(1, p, n_samples)

p_defect   = 1 - p             
mean_th    = p                  
var_th     = p * (1 - p)       

mean_emp   = np.mean(samples)
var_emp    = np.var(samples)

print("=" * 45)
print("   Microchip Quality-Control — Bernoulli")
print("=" * 45)
print(f"  p (pass probability)     : {p}")
print(f"  P(X=0) — Defect prob     : {p_defect:.4f}")
print(f"  E[X]   — Theoretical mean: {mean_th:.4f}")
print(f"  Var(X) — Theoretical var : {var_th:.4f}")
print("-" * 45)
print(f"  Empirical Mean           : {mean_emp:.4f}")
print(f"  Empirical Variance       : {var_emp:.4f}")
print("=" * 45)
```

```python
counts = [np.sum(samples == 0), np.sum(samples == 1)]
bars = plt.bar([0, 1], counts, color=['#e74c3c', '#2ecc71'],
               edgecolor='black', width=0.4)

for bar, cnt in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 10,
             f'{cnt}', ha='center', fontsize=11, fontweight='bold')

plt.title("Bernoulli Distribution — Microchip Quality Control (p = 0.85)", fontsize=13)
plt.xlabel("Outcome  (0 = Fail / Defect,  1 = Pass / Success)")
plt.ylabel("Frequency")
plt.xticks([0, 1], ['0 — Fail (Defect)', '1 — Pass (Success)'])
plt.ylim(0, max(counts) + 80)
plt.tight_layout()
plt.show()
```
