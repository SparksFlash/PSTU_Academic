### Lab 19: Emergency Room (ER) Patient Flow: Simulate the triage, treatment, and discharge of patients in an
ER to minimize waiting times and allocate resources

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Parameters
n_patients = 100
n_doctors  = 3

# Simulate patient arrivals and triage
arrival_times = np.cumsum(np.random.exponential(10, n_patients))
triage_done   = arrival_times + np.random.exponential(5, n_patients)
priorities    = np.random.choice(["Critical","Serious","Minor"],
                                  n_patients, p=[0.2, 0.3, 0.5])
treat_avg     = {"Critical": 30, "Serious": 20, "Minor": 10}

# Assign doctors and compute wait & treatment times
doctor_free = np.zeros(n_doctors)
wait, treat = [], []

for i in range(n_patients):
    d     = np.argmin(doctor_free)
    start = max(triage_done[i], doctor_free[d])
    t     = np.random.exponential(treat_avg[priorities[i]])
    doctor_free[d] = start + t
    wait.append(start - triage_done[i])
    treat.append(t)

wait, treat = np.array(wait), np.array(treat)

# Print Results
print(f"{'='*50}")
print(f"  ER Simulation — {n_patients} Patients, {n_doctors} Doctors")
print(f"{'='*50}")
print(f"  Avg Wait Time  : {wait.mean():.2f} min")
print(f"  Max Wait Time  : {wait.max():.2f} min")
print(f"  Avg Treat Time : {treat.mean():.2f} min")
print(f"{'─'*50}")
for p in ["Critical","Serious","Minor"]:
    m = priorities == p
    print(f"  {p:8s} | Patients: {m.sum():3d} | "
          f"Avg Wait: {wait[m].mean():.1f} min | Avg Treat: {treat[m].mean():.1f} min")
print(f"{'='*50}")

# Plots
labels = ["Critical", "Serious", "Minor"]
colors = ['#e74c3c', '#f39c12', '#2ecc71']
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

axes[0].hist(wait, bins=20, color='#3498db', edgecolor='black')
axes[0].set(title="Waiting Time Distribution", xlabel="Wait (min)", ylabel="Patients")

axes[1].bar(labels, [np.sum(priorities == p) for p in labels], color=colors, edgecolor='black')
axes[1].set(title="Patients by Priority", xlabel="Priority", ylabel="Count")

axes[2].bar(labels, [wait[priorities == p].mean() for p in labels], color=colors, edgecolor='black')
axes[2].set(title="Avg Wait by Priority", xlabel="Priority", ylabel="Avg Wait (min)")

plt.tight_layout()
plt.show()
```
